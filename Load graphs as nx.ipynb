```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python

def load(name):
    assert name in ["ego","grid","sbm","planar"]

    A_train, A_val, A_test = np.load("dataset/"+name+"/"+name+"_train_val_test.npy")

    train, val, test = [],[],[]

    for A in A_train:
        train.append(nx.from_numpy_array(A))
    for A in A_val:
        val.append(nx.from_numpy_array(A))
    for A in A_test:
        test.append(nx.from_numpy_array(A))
        
    return train,val,test
```

```python
train,val,test = load("planar")
print(len(train),len(val),len(test))
```

```python
nx.draw(train[0],node_size=10)
plt.show()
```
